```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

HOME = os.path.expanduser('~')
cust = pd.read_csv(HOME + '/temp/santander/train.csv', encoding='latin-1')
cust.info()
```

```python
cust.head()
```

```python
cust['TARGET'].value_counts()
```

```python
print('불만족 고객 비율: {0} %'.format(np.round(cust['TARGET'].value_counts()[1] / cust['TARGET'].count() * 100)))
```

```python
cust.describe()
```

```python
cust['var3'].replace(-999999, 2, inplace=True)
cust.drop('ID', axis=1, inplace=True)

x_features = cust.iloc[:, :-1]
y_labels = cust.iloc[:, -1]

print('피처 데이터 shape:{0}'.format(x_features.shape))
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_features, y_labels,
                                                   test_size=0.2, random_state=0)

train_cnt = y_train.count()
test_cnt = y_test.count()

print('학습 세트 Shape: {0}, 테스트 세트 Shape: {1}'.format(x_train.shape, x_test.shape))
```

```python
print('학습 세트 레이블 값 분포 비율')
print(y_train.value_counts() / train_cnt)
```

```python
print('테스트 세트 레이블 값 분포 비율')
print(y_test.value_counts() / test_cnt)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score

xgb = XGBClassifier(n_estimators=500, random_state=156)

xgb.fit(x_train, y_train, early_stopping_rounds=100,
       eval_metric='auc', eval_set=[(x_train, y_train), (x_test, y_test)])

score = roc_auc_score(y_test, xgb.predict_proba(x_test)[:, 1], average='macro')
score
```

```python
from sklearn.model_selection import GridSearchCV

xgb = XGBClassifier(n_estimators=100)

params = {
    'max_depth': [5, 7],
    'min_child_weight': [1, 3],
    'colsample_bytree': [0.5, 0.75]
}

grid = GridSearchCV(xgb, param_grid=params, cv=3)
grid.fit(x_train, y_train, early_stopping_rounds=30, eval_metric='auc',
        eval_set=[(x_train, y_train), (x_test, y_test)])

grid.best_params_
```

```python
score = roc_auc_score(y_test, grid.predict_proba(x_test)[:, 1], average='macro')
score
```

```python
xgb = XGBClassifier(n_estimators=1000, random_state=156, learning_rate=0.02,
                   max_depth=7, min_child_weight=1, colsample_bytree=0.75, reg_alpha=0.03)

xgb.fit(x_train, y_train, early_stopping_rounds=200,
       eval_metric='auc', eval_set=[(x_train, y_train), (x_test, y_test)])

score = roc_auc_score(y_test, xgb.predict_proba(x_test)[:, 1], average='macro')
score
```

```python
from xgboost import plot_importance
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(1, 1, figsize=(10, 8))
plot_importance(xgb, ax=ax, max_num_features=20, height=0.4)
```

```python
from lightgbm import LGBMClassifier

lgbm = LGBMClassifier(n_estimators=500)

evals = [(x_test, y_test)]
lgbm.fit(x_train, y_train, early_stopping_rounds=100, eval_metric='auc',
        eval_set=evals, verbose=True)

score = roc_auc_score(y_test, lgbm.predict_proba(x_test)[:, 1], average='macro')
score
```

```python
lgbm = LGBMClassifier(n_estimators=200)

params = {
    'num_leaves': [32, 64],
    'max_depth': [128, 160],
    'min_child_samples': [60, 100],
    'subsample': [0.8, 1]
}

grid = GridSearchCV(lgbm, param_grid=params, cv=3)
grid.fit(x_train, y_train, early_stopping_rounds=30, eval_metric='auc',
        eval_set=[(x_train, y_train), (x_test, y_test)])

grid.best_params_
```

```python
score = roc_auc_score(y_test, grid.predict_proba(x_test)[:, 1], average='macro')
score
```

```python
lgbm = LGBMClassifier(n_estimators=1000, num_leaves=32, sumbsample=0.8, min_child_samples=100,
                     max_depth=128)

evals = [(x_test, y_test)]

lgbm.fit(x_train, y_train, early_stopping_rounds=100, eval_metric='auc',
        eval_set=evals, verbose=True)

score = roc_auc_score(y_test, lgbm.predict_proba(x_test)[:, 1], average='macro')
score
```
